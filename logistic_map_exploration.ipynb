```python
# Runs on Python 3.9.10

import numpy as np
import matplotlib.pyplot as plt
```

```python
def log_map(r,x):
  return r*x*(1-x)
```

```python

N = 10000
M = 5000
B = 500
C = 1000

r_start = 3.3
r_end = 3.8

oplot = np.zeros((B,C))

for i in range(C):
  r = r_start + (r_end-r_start)*i/C

  xn = 0.2
  #print(r)
  y_dirty = np.zeros((N))
  for j in range(N):
    y_dirty[j] = xn
    xn = log_map(r,xn)
  y = y_dirty[-M:]
  h = np.histogram(y, bins = np.linspace(0,1,B+1))[0]
  h = h / max(h)
  oplot[...,i] = h
```

```python
plt.imshow(np.flipud(oplot),extent=(r_start,r_end,0,1),aspect = "auto")
plt.show()
```
